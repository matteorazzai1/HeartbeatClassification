# Data Cleaning and Imbalanced Class

According to some related papers I decided to not consider the two minority class F and Q. The class Q (unknown) is discarded because doesn't give us a response regarding the beat and the class F (Fusion beat between Normal beat and Ventricular beat) regards a beat that don't give us a response regarding the normality of the beat, and it occurs often when there is a treatment with a pacemaker, so, reading some paper on this argument, I decide to drop this minority class for this experiment.

Paper: Secure Edge Computing, chapter 9, by M. Ahmed, P. Haskell-Downland

```python
#in this way the record regarding the two classes are dropped
df_cardio=df_cardio[df_cardio['type']!= 'Q']
df_cardio=df_cardio[df_cardio['type']!= 'F']
```

I use OrdinalEncoder, because some of the models used work better with numerical variables, so I convert the 'type' features with this relation:

    - N     --> 0.0
    - SVEB  --> 1.0
    - VEB   --> 2.0

```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
enc=OrdinalEncoder()
X=df_cardio.drop('type',axis=1)
y=df_cardio['type']
encoded_class=enc.fit_transform(y.values.reshape(-1,1))

df_cardio['type']=encoded_class
df_cardio
```

# Data Normalization

An atrial premature beats usually means the presence of a Supraventricular anomalies, and a premature ventricular contraction beats usually means the presence of a Supraventricular anomalies.

This means that the RR-interval features (0_pre-RR,0_post-RR,1_pre-RR,1_post-RR) are very important for the detection of the anomalies. 

As is said in the following paper:C.-C. Lin and C.-M. Yang, ‘‘Heartbeat classification using normalized RR
intervals and morphological features,’’, the presence of arrhythmic heartbeats would shorten or prolong the previous or post-RR intervals. However, the RR interval features were directly related with the heart rate.
The inconsistency of the heart rates between ECG recordings would reduce the classification performance of the RR interval features.

For avoid this behaviour the RR interval features were normalized, using the mean and the standard deviation of the values of that specific features for a specific patient.

```python
groups=df_cardio.groupby('patient')
groups
```

```python
for name,patient in groups:
    #print(patient)
    df_patient=pd.DataFrame(patient)
    
    RR_pre0=df_patient['0_pre-RR']
    RR_pre1=df_patient['1_pre-RR']
    RR_post0=df_patient['0_post-RR']
    RR_post1=df_patient['1_post-RR']
    
    RR_pre0=RR_pre0.to_numpy()
    RR_pre1=RR_pre1.to_numpy()
    RR_post0=RR_post0.to_numpy()
    RR_post1=RR_post1.to_numpy()
    
    meanRR_pre0=np.mean(RR_pre0)
    meanRR_pre1=np.mean(RR_pre1)
    meanRR_post0=np.mean(RR_post0)
    meanRR_post1=np.mean(RR_post1)
    
    std0= np.std(RR_pre0)
    std1=np.std(RR_pre1)
    std2=np.std(RR_post0)
    std3=np.std(RR_post1)
    
    RR_pre0_norm=(RR_pre0-meanRR_pre0)/std0
    RR_pre1_norm=(RR_pre1-meanRR_pre1)/std1
    RR_post0_norm=(RR_post0-meanRR_post0)/std2
    RR_post1_norm=(RR_post1-meanRR_post1)/std3
    
    print("mean pre_0:"+str(np.mean(RR_pre0_norm))) 
    print("std pre_0:"+str(np.std(RR_pre0_norm))) 
    
    print("mean pre_1:"+str(np.mean(RR_pre1_norm))) 
    print("std pre_1:"+str(np.std(RR_pre1_norm)))
    
    print("mean post_0:"+str(np.mean(RR_post0_norm))) 
    print("std post_0:"+str(np.std(RR_post0_norm)))
    
    print("mean post_1:"+str(np.mean(RR_post1_norm))) 
    print("std post_1:"+str(np.std(RR_post1_norm)))
     
    
    df_patient['0_pre-RR']=RR_pre0_norm
    df_patient['1_pre-RR']=RR_pre1_norm
    df_patient['0_post-RR']=RR_post0_norm
    df_patient['1_post-RR']=RR_post1_norm
    
    df_cardio[df_cardio['patient']==name]=df_patient
```
